```python
df = pd.read_excel('Project_Data.xlsx')

# Store the maximum values row separately
max_row = df.iloc[-1]

# Drop the last row and the empty row before it
df = df.dropna().iloc[:-1]

max_row
```

# PART 1

```python
# Convert columns Q1 through Total to numeric
columns_to_convert = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Total'] 
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')
df
```

```python
# Replace NaN values with column-wise median
df[columns_to_convert] = df[columns_to_convert].fillna(df[columns_to_convert].median())
df
```

```python
# Replace values greater than column maximum or less than 0 with column-wise median
for column in columns_to_convert:
    column_max = df[column].max()
    column_median = df[column].median()
    df.loc[(df[column] > column_max) | (df[column] < 0), column] = column_median
df
```

```python
# Calculate the sum of Q1 to Q5 for each row and update the Total column
df['Total'] = df['Q1'] + df['Q2'] + df['Q3'] + df['Q4'] + df['Q5']

# Find the maximum value in the Total column
max_total = df['Total'].max()

# Drop rows where the Total value is greater than max_total
df = df[df['Total'] <= max_total]
df
```

outlier

```python
# Detect outliers using IQR method with a threshold of 2.5
for column in columns_to_convert:
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    threshold = 2.5 * iqr
    df.loc[(df[column] < q1 - threshold) | (df[column] > q3 + threshold), column] = df[column].median()
df
```

```python
df_preprocessed = df.copy()
```

preprocess finished

# PART 2

move mean to 0 and make std 1

```python
# Select columns Q1 to Total
columns_to_scale = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Total']

df_scaled = df_preprocessed.copy()
# Calculate the mean and standard deviation of each column
column_means = df_scaled[columns_to_scale].mean()
column_stds = df_scaled[columns_to_scale].std()

# Subtract the column means from each column
df_scaled[columns_to_scale] -= column_means

# Divide each column by its standard deviation
df_scaled[columns_to_scale] /= column_stds

df_scaled
```

# PART 3

```python
df_reduced = df_scaled.copy()
# Define the columns to be dropped
columns_to_drop = []


# Loop twice for each column to be dropped
for iteration in range(2):
    # Get all columns except 'Unnamed: 0' and 'Total'
    all_columns = [column for column in df_reduced.columns if column not in ['Unnamed: 0', 'Total']]

    print(f"Iteration {iteration+1}:")
    
    # Calculate the correlation matrix between columns Q1 to Q5
    corr_matrix = df_reduced[all_columns].corr()
    print("Correlation Matrix:")
    print(corr_matrix)
    
    # Find the column pair with the highest covariance
    max_covariance = 0
    max_cov_columns = None
    for i in range(len(corr_matrix.columns)):
        for j in range(i+1, len(corr_matrix.columns)):
            covariance = corr_matrix.iloc[i, j]
            if covariance > max_covariance:
                max_covariance = covariance
                max_cov_columns = (corr_matrix.columns[i], corr_matrix.columns[j])
    print("Max Correlation Columns:", max_cov_columns)
    
    # Calculate the covariance of the two columns with the label column
    label_column = 'Total'  # Replace 'Total' with the actual label column name
    cov1 = df_reduced[max_cov_columns[0]].cov(df_reduced[label_column])
    cov2 = df_reduced[max_cov_columns[1]].cov(df_reduced[label_column])
    print(f"Covariance of {max_cov_columns[0]} with {label_column}: {cov1}")
    print(f"Covariance of {max_cov_columns[1]} with {label_column}: {cov2}")
    
    # Determine the column to be dropped based on the covariance with the label column
    column_to_drop = max_cov_columns[0] if cov1 < cov2 else max_cov_columns[1]
    columns_to_drop.append(column_to_drop)
    print("Column to Drop:", column_to_drop)
    
    # Drop the selected column from the DataFrame if it exists
    if column_to_drop in df_reduced.columns:
        df_reduced = df_reduced.drop(column_to_drop, axis=1)
        print("Dropped Column:", column_to_drop)
    
df_reduced
```
